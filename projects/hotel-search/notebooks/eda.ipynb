```python
import pandas as pd

filename = "https://storage.googleapis.com/superlinked-recipes/hotels-search/dataset.jsonl"

df = pd.read_json(filename, lines=True)

df.sample(5, random_state=42)
```

```python
pd.DataFrame(
    {
        "n_hotels": df.country.value_counts(),
        "n_cities": df.groupby("country")["city"].nunique(),
    }
).sort_values(["n_hotels", "n_cities"], ascending=False)
```

```python
df[["price", "rating", "rating_count"]].describe()
```

```python
pd.DataFrame(
    {
        "chars": df.description.str.len().describe(),
        "words": df.description.str.split().str.len().describe(),
    }
)
```

```python
columns_categorical = [
    "property_amenities",
    "room_amenities",
    "wellness_spa",
    "accessibility",
    "for_children",
]

for column in columns_categorical:
    print(f"{column} | # unique: {df[column].explode().nunique()}")
    n = 5
    print(f"Top {n}:", df[column].explode().value_counts()[:n].to_dict(), "\n")
```
